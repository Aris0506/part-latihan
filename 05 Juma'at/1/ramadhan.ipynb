```python
import duckdb
import pandas as pd

con = duckdb.connect()
df_check = con.sql("SELECT * FROM 'WA_Fn-UseC_-Telco-Customer-Churn.csv' LIMIT 5").df()
display(df_check)
```

```python
gender = con.sql("SELECT DISTINCT gender FROM df_check").to_df()
print(gender)
```
